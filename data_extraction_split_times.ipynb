```python
import PyPDF2
import pandas as pd
import re

# Open the PDF file
with open('2020 Olympic 1500m Finals Splits.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    num_pages = len(reader.pages)

    # Initialize lists to store athlete names and split times
    athlete_names = []
    split_times = {100 * i: [] for i in range(1, 16)}  # Split distances per 100m

    # Iterate through each page of the PDF

    page = reader.pages[0]
    text = page.extract_text()
    #print(text)

    # Split the text into lines
    lines = text.split('\n')
    
    athlete_data = []
    athlete_data.append(lines[53:87])
    athlete_data.append(lines[87:121])
    athlete_data.append(lines[121:156])
    athlete_data.append(lines[156:191])
    athlete_data.append(lines[191:226])
    athlete_data.append(lines[226:261])
    athlete_data.append(lines[261:295])
    athlete_data.append(lines[295:330])
    athlete_data.append(lines[330:364])
    athlete_data.append(lines[364:398])
    athlete_data.append(lines[398:432])
    athlete_data.append(lines[432:466])
    athlete_data.append(lines[466:500])

    #print(athlete_data)


    athlete_data_filtered = [[item for item in row if "PB" not in item] for row in athlete_data]

    final_ranks = [row[0] for row in athlete_data_filtered]
    athlete_names = [row[2] for row in athlete_data_filtered]
    nationalities = [row[3] for row in athlete_data_filtered]
    final_times = [row[4] for row in athlete_data_filtered]

    # Extract split ranks and times for each athlete
    split_ranks = []
    split_times = []

    for athlete_info in athlete_data_filtered:
        # Extract split ranks and times
        ranks = []
        times = []
        for i in range(6, len(athlete_info), 2):
            time = athlete_info[i]
            rank = athlete_info[i+1]
            times.append(time)
            ranks.append(int(rank[1:-1]))
        times.append(athlete_info[4])
        ranks.append(int(athlete_info[0]))

        split_ranks.append(ranks)
        split_times.append(times)

    # Create a dictionary with athlete information
    athletes_dict = {}
    for name, rank, nationality, time, rank_list, time_list in zip(athlete_names, final_ranks, nationalities, final_times, split_ranks, split_times):
        athletes_dict[name] = {
            'Final Rank': rank,
            'Nationality': nationality,
            'Final Time': time,
            'Split Ranks': rank_list,
            'Split Times': time_list
        }

    # Print the dictionary
    for name, info in athletes_dict.items():
        print(f"Athlete: {name}")
        print("Final Rank:", info['Final Rank'])
        print("Nationality:", info['Nationality'])
        print("Final Time:", info['Final Time'])
        print("Split Ranks:", info['Split Ranks'])
        print("Split Times:", info['Split Times'])
        print()

df = pd.DataFrame.from_dict(athletes_dict, orient='index')

# Save the DataFrame to a CSV file
df.to_excel('athletes_data.xlsx', index_label='Athlete')

print("CSV file saved successfully.")
```
