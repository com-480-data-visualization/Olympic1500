```python
import PyPDF2
import pandas as pd
import re

# Open the PDF file
with open('2016 Olympic 1500m Finals Splits.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    num_pages = len(reader.pages)

    # Initialize lists to store athlete names and split times
    athlete_names = []
    split_times = {100 * i: [] for i in range(1, 16)}  # Split distances per 100m

    # Iterate through each page of the PDF

    page = reader.pages[0]
    text = page.extract_text()
    #print(text)

    # Split the text into lines
    lines = text.split('\n')    
    athlete_data = []
    start_index = 19
    end_index = 22
    while end_index <= len(lines):
        athlete_data.append(lines[start_index:end_index])
        start_index += 3
        end_index += 3
    # Extracting final ranks, athlete names, nationalities, and final times
    final_ranks = []
    athlete_names = []
    nationalities = []
    final_times = []

    for data in athlete_data:
        parts = data[0].split()
        final_ranks.append(int(parts[0]))
        athlete_names.append(" ".join(parts[2:4]))
        nationalities.append(parts[4])
        final_times.append(parts[5])

    # Extracting split ranks and split times
    split_ranks = []
    split_times = []

    for data in athlete_data:
        splitsAndRanks = data[1].split() + data[2].split()

        # Extracting splits and ranks
        times = []
        ranks = []

        for i in range(0, len(splitsAndRanks), 2):
            times.append(splitsAndRanks[i])
            ranks.append(int(splitsAndRanks[i+1][1:-1]))
        ranks.append(int(data[0].split()[0]))
        times.append(data[0].split()[5])
        split_times.append(times)
        split_ranks.append(ranks)

    # Printing the extracted information
    print("Final Ranks:", final_ranks)
    print("Athlete Names:", athlete_names)
    print("Nationalities:", nationalities)
    print("Final Times:", final_times)
    print("Split Ranks:", split_ranks)
    print("Split Times:", split_times)

    extracted_data = {
        'Final Rank': final_ranks,
        'Athlete Name': athlete_names,
        'Nationality': nationalities,
        'Final Time': final_times,
        'Split Ranks': split_ranks,
        'Split Times': split_times
    }

    # Convert dictionary to DataFrame
    df = pd.DataFrame(extracted_data)

    # Save DataFrame to Excel
    df.to_excel('2016_extracted_splits.xlsx', index=False)

    print("Data saved to Excel successfully.")
```
